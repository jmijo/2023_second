# 결측치 제거

```python
sur1 = sur.copy()
```

```python
sur1.isnull().sum()
```

```python
sur1 = sur1.dropna(axis=0)
```

```python
sur1.isnull().sum()
```

```python
x1 = sur1[['ta', 'td', 'hm', 'ws', 'rn', 're', 'si', 'ss', 'q']]
y1 = sur1['ts']
```

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, train_size = 0.7 , random_state = 1234)
```

```python
dum1_train = pd.get_dummies(x1_train['q'], prefix='q')
dum1_train.head()
```

```python
dum1_test = pd.get_dummies(x1_test['q'], prefix='q')
dum1_test.head()
```

```python
x1_train.drop('q', axis = 1, inplace = True)
x1_test.drop('q', axis = 1, inplace = True)
```

```python
print(x1_train.shape)
print(dum1_train.shape)
print(x1_test.shape)
print(dum1_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x1_train = scaler.fit_transform(x1_train)
x1_test = scaler.fit_transform(x1_test)
```

```python
x1_train = np.concatenate((x1_train, np.array(dum1_train)), axis = 1)
x1_test = np.concatenate((x1_test,  np.array(dum1_test)), axis = 1)

print(x1_train[:3])
print("")
print(x1_test[:3])
```

```python
print(x1_train[:3])
print("")
print(x1_test[:3])
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
# 전진 단계별 선택법
import statsmodels.api as sm
df = pd.DataFrame(x1_train, columns = [['ta','td','hm','ws','rn','re','si','ss','q_1','q_2','q_3','q_4']])
variables = df.columns.tolist()  ## 설명 변수 리스트

y = y1_train.values ## 반응 변수
selected_variables2 = [] ## 선택된 변수들
sl_enter = 0.05 # 기준 유의확률
sl_remove = 0.05

sv_per_step = [] ## 각 스텝별로 선택된 변수들
adjusted_r_squared = [] ## 각 스텝별 수정된 결정계수
steps = [] ## 스텝
step = 0
while len(variables) > 0:
    remainder = list(set(variables) - set(selected_variables2))
    pval = pd.Series(index=remainder) ## 변수의 p-value
    ## 기존에 포함된 변수와 새로운 변수 하나씩 돌아가면서
    ## 선형 모형을 적합한다.
    for col in remainder:
        X = df[selected_variables2+[col]]
        X = sm.add_constant(X)
        model = sm.OLS(y,X).fit()
        pval[col] = model.pvalues[col]

    min_pval = pval.min()
    if min_pval < sl_enter: ## 최소 p-value 값이 기준 값보다 작으면 포함
        selected_variables2.append(pval.idxmin())
        ## 선택된 변수들에대해서
        ## 어떤 변수를 제거할지 고른다.
        while len(selected_variables2) > 0:
            selected_X = df[selected_variables2]
            selected_X = sm.add_constant(selected_X)
            selected_pval = sm.OLS(y,selected_X).fit().pvalues[1:] ## 절편항의 p-value는 뺀다
            max_pval = selected_pval.max()
            if max_pval >= sl_remove: ## 최대 p-value값이 기준값보다 크거나 같으면 제외
                remove_variable = selected_pval.idxmax()
                selected_variables2.remove(remove_variable)
            else:
                break

        step += 1
        steps.append(step)
        adj_r_squared = sm.OLS(y,sm.add_constant(df[selected_variables2])).fit().rsquared_adj
        adjusted_r_squared.append(adj_r_squared)
        sv_per_step.append(selected_variables2.copy())
    else:
        break
```

```python
print(selected_variables2)
```

```python
# 단순회귀
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x1_train, y1_train)

print(lr.score(x1_train, y1_train))
print(lr.score(x1_test, y1_test))
```

```python
plt.figure(figsize=(15,7))
plt.plot(lr.predict(x1_test[:100]), label='predict')
plt.plot(y1_test[:100].values.reshape(-1,1), label = 'real')
plt.legend()
```

```python
#MAE
pred_lr = lr.predict(x1_test)
print(mean_absolute_error(y1_test, pred_lr))

#MSE

print(mean_squared_error(y1_test, pred_lr))

#RMSE
MSE = mean_squared_error(y1_test, pred_lr)
print(np.sqrt(MSE))
```

```python
# 의사결정나무
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x1_train, y1_train)

print(dtr.score(x1_train, y1_train))
print(dtr.score(x1_test, y1_test))
```

```python
plt.figure(figsize=(15,7))
plt.plot(dtr.predict(x1_test[:100]), label='predict')
plt.plot(y1_test[:100].values.reshape(-1,1), label = 'real')
plt.legend()
```

```python
pred_dtr = dtr.predict(x1_test)

#MAE
print(mean_absolute_error(y1_test, pred_dtr))

#MSE

print(mean_squared_error(y1_test, pred_dtr))

#RMSE
MSE = mean_squared_error(y1_test, pred_dtr)
print(np.sqrt(MSE))

#MAPE
print(mean_absolute_percentage_error(y1_test, pred_dtr))
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x1_train, y1_train)

print(rfr.score(x1_train, y1_train))
print(rfr.score(x1_test, y1_test))
```

```python
plt.figure(figsize=(15,7))
plt.plot(rfr.predict(x1_test[:100]), label='predict')
plt.plot(y1_test[:100].values.reshape(-1,1), label = 'real')
plt.legend()
```

```python
pred_rfr = rfr.predict(x1_test)

#MAE
print(mean_absolute_error(y1_test, pred_rfr))

#MSE

print(mean_squared_error(y1_test, pred_rfr))

#RMSE
MSE = mean_squared_error(y1_test, pred_rfr)
print(np.sqrt(MSE))
```

```python
plt.figure(figsize=(15,7))
plt.plot(xgb.predict(x1_test[:100]), label='predict')
plt.plot(y1_test[:100].values.reshape(-1,1), label = 'real')
plt.legend()
```
